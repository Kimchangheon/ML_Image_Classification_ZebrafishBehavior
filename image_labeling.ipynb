```python
dic = {"control" : 0 ,  "sc" : 0 ,'10ppm' : 0 , '100ppm' : 0,'200ppm' : 0, 'BIT0.01' : 0, 'BIT0.1' : 0, 'BIT1' : 0, 'pb0.1':0 ,'Pb0.5':0, 'bpa0.01' : 0, 'bpa0.1' : 0, 'bpa1' : 0}
```

```python
import os, shutil
# base_dir = "F:/UOSEST/P4_CNN_ZebraFish/All_alive_labeling"
base_dir = "./All_alive_labeling"
os.mkdir(base_dir)
for i in dic.keys() :
    os.mkdir(os.path.join(base_dir, i))
```

```python
for i in file_list : 
    if '100ppm' in i :
        dic['100ppm'] +=1
        shutil.copy(path_dir+i, base_dir+'/100ppm/'+i)
        
    elif '10ppm' in i :
        dic['10ppm'] +=1
        shutil.copy(path_dir+i, base_dir+'/10ppm/'+i)
        
    elif '200ppm' in i :
        dic['200ppm'] +=1
        shutil.copy(path_dir+i, base_dir+'/200ppm/'+i)
        
    elif 'BIT0.01' in i :
        dic['BIT0.01'] +=1
        shutil.copy(path_dir+i, base_dir+'/BIT0.01/'+i)
        
    elif 'BIT0.1' in i:
        dic['BIT0.1'] +=1
        shutil.copy(path_dir+i, base_dir+'/BIT0.1/'+i)

        
    elif 'BIT1' in i :
        dic['BIT1'] += 1
        shutil.copy(path_dir+i, base_dir+'/BIT1/'+i)
        
    elif 'pb0.1' in i :
        dic['pb0.1'] += 1
        shutil.copy(path_dir+i, base_dir+'/pb0.1/'+i)
        
    elif (('Pb0.5' in i) | ('Pb-0.5' in i)):
        dic['Pb0.5'] += 1
        shutil.copy(path_dir+i, base_dir+'/Pb0.5/'+i)
        
    elif(('sc' in i) | ('SC' in i) | ('bpasol' in i)) : 
        dic['sc'] +=1
        shutil.copy(path_dir+i, base_dir+'/sc/'+i)
        
    elif 'bpa0.1' in i :
        dic['bpa0.1'] +=1
        shutil.copy(path_dir+i, base_dir+'/bpa0.1/'+i)
        
    elif 'bpa0.01' in i :
        dic['bpa0.01'] +=1
        shutil.copy(path_dir+i, base_dir+'/bpa0.01/'+i)

        
    elif 'bpa1' in i :
        dic['bpa1'] +=1
        shutil.copy(path_dir+i, base_dir+'/bpa1/'+i)

        
    else : 
        dic['control'] +=1
        shutil.copy(path_dir+i, base_dir+'/control/'+i)
```

```python
for i  in dic :
    print(i +" : "+ str(dic[i]))
```
